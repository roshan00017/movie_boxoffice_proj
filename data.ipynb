### Box office parsing into integers


```python
df['WorldwideBox Office'] = df['WorldwideBox Office'].astype(str)
df['DomesticBox Office'] = df['DomesticBox Office'].astype(str)
df['InternationalBox Office'] = df['InternationalBox Office'].astype(str)


print(df.dtypes)
```

```python
df['WorldwideBox Office'] = df['WorldwideBox Office'].apply(lambda x: x.replace(',','').replace('$',''))
df['DomesticBox Office'] = df['DomesticBox Office'].apply(lambda x: x.replace(',','').replace('$',''))
df['InternationalBox Office'] = df['InternationalBox Office'].apply(lambda x: x.replace(',','').replace('$',''))
```
