```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Modified_data.csv')
```

```python
df.head()
```

```python
df.rename(columns={'Unnamed: 0':'Sn'}, inplace=True)
```

```python
df.head()
```

```python
df.nunique()
```

```python
df['Year'].unique()
```

```python
cordata = df.drop(['Rank','Year','Sn'],axis=1)
```

```python
cordata
```

## relationship analysis

```python
corelation = cordata.corr()
```

```python
sns.heatmap(corelation,xticklabels=corelation.columns,annot=True)
```

### Correlaton between Domestic , Internatonal And World Wide Box Office

```python
sns.pairplot(cordata)
```

```python
sns.relplot(x='DomesticBox Office', y='InternationalBox Office', hue='Year', kind='scatter', data=df)
```

#### It certainly shows the positive relationship between Domestic and International Box Offices. However, according to scatterplot, it is somewhat uncertain.

```python
year_num = df['Year'][:400].value_counts()
```

```python
chart = sns.barplot(x=year_num.index,y=year_num)
chart.set_xticklabels(chart.get_xticklabels(),rotation=90)
plt.show()
```

Above Barplot shows that in 2018, movie released on 2018 had great success in the box office .
